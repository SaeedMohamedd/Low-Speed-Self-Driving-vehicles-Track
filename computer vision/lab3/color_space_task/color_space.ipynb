```python
img = cv2.imread('001.jpg')
ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
y = ycrcb[:,:,0]*1.25
cr = ycrcb[:,:,1]
cb = ycrcb[:,:,2]

BGR_OUT = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)
cv2.imwrite("task2_out.jpg", BGR_OUT)

print("image save")
```

```python
cmapA = matplotlib.colors.LinearSegmentedColormap.from_list("", ["white","red"])
cmapB = matplotlib.colors.LinearSegmentedColormap.from_list("", ["blue","white"])

# Visualize the individual color channels
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25,20))
ax1.set_title('Original Image')
ax1.imshow(img)
ax2.set_title('Y channel')
ax2.imshow(y, cmap='gray')
ax3.set_title('Cr channel')
ax3.imshow(cr, cmap=cmapA)
ax4.set_title('Cb channel')
ax4.imshow(cb, cmap=cmapB);
```

```python
y =y * 1.25

# Visualize the individual color channels
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(25,20))
ax1.set_title('Original Image')
ax1.imshow(img)
ax2.set_title('Y channel')
ax2.imshow(y, cmap='gray')
ax3.set_title('Cr channel')
ax3.imshow(cr, cmap=cmapA)
ax4.set_title('Cb channel')
ax4.imshow(cb, cmap=cmapB);
```
