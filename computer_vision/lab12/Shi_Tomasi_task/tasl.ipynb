## Applay SHI_Tomasi on chess img

```python
import numpy as np
import cv2
from datetime import datetime
import time
import random as rng
import matplotlib.pyplot as plt
%matplotlib inline

img_resize = cv2.imread('chess.png')
img = cv2.resize(img_resize, (500, 500))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

def nothing(x):
    pass

cv2.namedWindow('shi_tomasi', cv2.WINDOW_NORMAL)

cv2.createTrackbar('maxCorners', 'shi_tomasi', 1, 500, nothing)
cv2.createTrackbar('qualityLevel', 'shi_tomasi', 1, 50, nothing)
cv2.createTrackbar('minDistance', 'shi_tomasi', 1, 500, nothing)
#cv2.createTrackbar('blocksize', 'shi_tomasi', 1, 20, nothing)
#cv2.createTrackbar('gradientSize', 'shi_tomasi', 1, 20, nothing)
#cv2.createTrackbar('k', 'shi_tomasi', 1, 20, nothing)


while 1:
    image= gray.copy()
    img2=img.copy()
    maxCorners = cv2.getTrackbarPos('maxCorners','shi_tomasi')
    qualityLevel = cv2.getTrackbarPos('qualityLevel','shi_tomasi') /100
    minDistance = cv2.getTrackbarPos('minDistance','shi_tomasi')
    #blockSize = cv2.getTrackbarPos('blocksize','shi_tomasi')
    #gradientSize = cv2.getTrackbarPos('gradientSize','shi_tomasi')
    #k = cv2.getTrackbarPos('k','shi_tomasi') /100
    if maxCorners==0:
        maxCorners=1
    #if blockSize == 0 :
    #    blockSize=1
    #dt1 = datetime.now()
    #dt1.microsecond
    start = time.time()
    corners = cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance)
    end = time.time()
    print((end - start)*1000) # to convert to millisecond
    corners = np.int0(corners)
    # draw red color circles on all corners
    for i in corners:
        x, y = i.ravel()
        cv2.circle(img, (x, y), 3, (255, 0, 0), -1)
    
    cv2.imshow('shi_tomasi', img)

    k = cv2.waitKey(1)
    if k == 27:
        break
    if k == 115:
        cv2.imwrite("chess_edges.png", img)

cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
from datetime import datetime
import time
import random as rng
import matplotlib.pyplot as plt
%matplotlib inline

img_resize = cv2.imread('saved_img.png')
imgg = cv2.resize(img_resize, (500, 500))
gray = cv2.cvtColor(imgg, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

def nothing(x):
    pass

cv2.namedWindow('shi_tomasi', cv2.WINDOW_NORMAL)

cv2.createTrackbar('maxCorners', 'shi_tomasi', 1, 500, nothing)
cv2.createTrackbar('qualityLevel', 'shi_tomasi', 1, 50, nothing)
cv2.createTrackbar('minDistance', 'shi_tomasi', 1, 500, nothing)
#cv2.createTrackbar('blocksize', 'shi_tomasi', 1, 20, nothing)
#cv2.createTrackbar('gradientSize', 'shi_tomasi', 1, 20, nothing)
#cv2.createTrackbar('k', 'shi_tomasi', 1, 20, nothing)


while 1:
    image= gray.copy()
    img=imgg.copy()
    maxCorners = cv2.getTrackbarPos('maxCorners','shi_tomasi')
    qualityLevel = cv2.getTrackbarPos('qualityLevel','shi_tomasi') /100
    minDistance = cv2.getTrackbarPos('minDistance','shi_tomasi')
    #blockSize = cv2.getTrackbarPos('blocksize','shi_tomasi')
    #gradientSize = cv2.getTrackbarPos('gradientSize','shi_tomasi')
    #k = cv2.getTrackbarPos('k','shi_tomasi') /100
    if maxCorners==0:
        maxCorners=1
    if qualityLevel == 0 :
        qualityLevel=.01
    #dt1 = datetime.now()
    #dt1.microsecond
    start = time.time()
    corners = cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance)
    end = time.time()
    print((end - start)*1000) # to convert to millisecond
    corners = np.int0(corners)
    # draw red color circles on all corners
    for i in corners:
        x, y = i.ravel()
        cv2.circle(img, (x, y), 3, (255, 0, 0), -1)
    
    cv2.imshow('shi_tomasi', img)

    k = cv2.waitKey(1)
    if k == 27:
        break
    if k == 115:
        cv2.imwrite("save_img_edges.png", img)

cv2.destroyAllWindows()
plt.imshow(img)
```

## demo code

```python
# Python program to illustrate 
# corner detection with 
# Shi-Tomasi Detection Method
    
# organizing imports 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
  
# path to input image specified and  
# image is loaded with imread command
img = cv2.imread('chess.png')
  
# convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
# Shi-Tomasi corner detection function
# We are detecting only 100 best corners here
# You can change the number to get desired result.
corners = cv2.goodFeaturesToTrack(gray_img, 100, 0.01, 10)
  
# convert corners values to integer
# So that we will be able to draw circles on them
corners = np.int0(corners)
  
# draw red color circles on all corners
for i in corners:
    x, y = i.ravel()
    cv2.circle(img, (x, y), 3, (255, 0, 0), -1)
  
# resulting image
plt.imshow(img)
  
# De-allocate any associated memory usage  
if cv2.waitKey(0) & 0xff == 27: 
    cv2.destroyAllWindows()
```
