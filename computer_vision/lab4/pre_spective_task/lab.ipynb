```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def SHOW(img):
    display(Image.fromarray(img[:,:,::-1]));
```

```python
img=cv2.imread('perspective02.jpg')
rows,cols = img.shape[:2]
res = cv2.resize(img,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)
#cv2.imshow("m",img)
```

```python
def nothing(x):
    pass

cv2.namedWindow('Image')

# create trackbars
cv2.createTrackbar('x', 'Image', 0, 1, nothing)

while 1:

    pts1 = np.float32(
        [
            [580,  334], [694,  344],
            [50, 640], [1100, 640]
        ]
    )

    pts2 = np.float32(
        [
            [580,   0], [694,  0],
            [580, 640], [694, 640]
        ]
    )

    M = cv2.getPerspectiveTransform(pts1 ,pts2)
    dst = cv2.warpPerspective(img, M, (cols,rows))

    
        
    for point in pts2:
        dst = cv2.circle(dst, (int(point[0]), int(point[1])), 1, (0, 0, 255), 2)

    x = cv2.getTrackbarPos('x', 'Image')
    
    out = []
    
    if (x == 1):
        out = cv2.addWeighted(img, 0.7, dst, 0.5, 0)
    else:
        for point in pts1:
            dst = cv2.circle(dst, (int(point[0]), int(point[1])), 1, (0, 255, 0), 2)
        out = dst

    cv2.imshow('Image', out)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break

cv2.destroyAllWindows()


SHOW(out);
```
