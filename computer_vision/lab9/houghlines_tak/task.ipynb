```python
import cv2
import numpy as np
img = cv2.imread('Building.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def nothing(x):
    pass

cv2.namedWindow('Hough')

cv2.createTrackbar('x', 'Hough', 0, 800, nothing)
cv2.createTrackbar('y', 'Hough', 0, 20, nothing)
cv2.createTrackbar('z', 'Hough', 0, 20, nothing)
edges = cv2.Canny(gray, 50, 200)

while 1:
    image = edges.copy()
    x = cv2.getTrackbarPos('x','Hough')
    min_lines_lenght=cv2.getTrackbarPos('y','Hough')
    max_lines_gap=cv2.getTrackbarPos('z','Hough')
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, x,min_lines_lenght,max_lines_gap)
    if lines is not None:
        for i in range(0, len(lines)):
            rho = lines[i][0][0]
            theta = lines[i][0][1]

            x0 = rho * np.cos(theta)
            y0 = rho * np.sin(theta)

            a = np.cos(theta)
            b = np.sin(theta)

            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            #cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)
            for x1,y1,x2,y2 in lines[i]:
                cv2.line(image,(x1,y1),(x2,y2),(255,0,0),10)
            #cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)
            color_edge = np.dstack((edges, edges,edges))
            #image=cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
            combination = cv2.addWeighted(color_edge[:,:,2], 0.8, image, 1, 0)  
    #edgesC = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    #image = cv2.addWeighted(image, 0.9, edgesC, 0.5, 0)
    cv2.imshow('Hough', image)
    k = cv2.waitKey(1)
    if k == 27:
        break

cv2.destroyAllWindows()
cv2.imwrite("building_img_edges_red.jpg", image)
```
